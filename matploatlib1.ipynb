```python
#two chart in single grapgh
# you can show here multple graph even 2,3 4,5 6 more what you think
z=x**2.16
plt.plot(x,y,color='k',linestyle='--',linewidth=3,label='1st')
plt.plot(x,z,color='y',linestyle=':',linewidth=3,label='2nd')
plt.title('This is Line Ploat')
plt.xlabel('X-Values')
plt.ylabel('Y-Values')
plt.legend() # this is use for show 1st and 2nd lable name that's called legend
```

```python
#multiple graphs but,diffrent chart in matrix fomr like (2,2) (3,1) (3,2) like this

#chart_1 start
plt.subplot(3,1,1) # row 3, column 1, 1 first chart
plt.plot(x,y,color='y',linestyle=':',linewidth=3)
plt.title('This is Line Ploat')
plt.xlabel('X-Values')
plt.ylabel('Y-Values')
#chart_1 end

#chart_2 start
plt.subplot(3,1,2) # row 3, column 1, 2 first chart
plt.plot(x,x**5.8,color='g',linestyle=':',linewidth=3)
plt.title('This is Line Ploat')
plt.xlabel('X-Values')
plt.ylabel('Y-Values')
#chart_2 end

#chart_3 start
plt.subplot(3,1,3) # row 3, column 1, 3 first chart
plt.plot(x,x**89.9,color='r',linestyle=':',linewidth=3)
plt.title('This is Line Ploat')
plt.xlabel('X-Values')
plt.ylabel('Y-Values')
#chart_3 end
```

# Bar Ploat

```python
price={'apple':120,'orange':70,'banana':40,'pear':50,'Kiwi':180}
```

```python
fruit=list(price.keys())
fruit
```

```python
cost=list(price.values())
cost
```

```python
fig=plt.figure()
fig.suptitle('bold figure suptitle', fontsize=14, fontweight='bold')
plt.bar(fruit,cost,color='c',edgecolor='k',width=0.5,align='edge')
plt.grid(True)
plt.title('Fruit Market')
plt.xlabel('fruit')
plt.ylabel('cost')
plt.show()
```

```python
fig=plt.figure()
fig.suptitle('bold figure suptitle', fontsize=14, fontweight='bold')
plt.barh(fruit,cost,color='r',edgecolor='k',height=0.5,align='edge')
plt.grid(True) 
plt.title('Fruit Market')
plt.xlabel('fruit')
plt.ylabel('cost')
plt.show()
```

# Scatter ploat

```python
sx=np.random.randint(10,51,200)
sy=np.random.randint(10,51,200)
```

```python
plt.scatter(sx,sy,color='g',marker='*') # you can use here marker=*,.,+,- anything symobol
plt.show()
```

```python
# three plot show or multiple
sx=np.random.randint(10,51,200)
sy=np.random.randint(10,51,200)
sy1=np.random.randint(20,78,200)
plt.scatter(sx,sy,color='g',marker='*') # you can use here marker= *, . ,+, -, anything symobol
plt.scatter(sx,sy1,color='r',marker='+') 
# add lengend
plt.legend(['sx','sy1'])
plt.show()
```

```python
plt.subplot(2,2,1)
plt.scatter(sx,sy,color='g',marker='+')
plt.title('1st ploat')
plt.subplot(2,2,4)
plt.scatter(sx,sy1,color='b',marker='+')
plt.title('2nd ploat')
```

# Pie chart and Donught Chart

```python
empl={'ravi':50000,'brijesh':55000,'suman':63000,'akash':25000,'mukesh':40000}
```

```python
emp_names=empl.keys()
emp_salary=empl.values()
```

```python
plt.pie(emp_salary,labels=emp_names,autopct='%.2f%%',colors=['r','g','b','c','#ff56'],radius=1.4) # you can use %d%%,%s%%
plt.show()
```

```python
#create doughnut
plt.pie(emp_salary,labels=emp_names,autopct='%.2f%%',colors=['r','g','b','c','#ff56'],radius=1.4) # you can use %d%%,%s%%
plt.pie([1],colors=['w'],radius=0.75)
plt.show()
```
