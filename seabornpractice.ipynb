```python
import seaborn as sns
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import warnings #to avoid warnings
warnings.filterwarnings('ignore')
```

```python
sns.get_dataset_names() # it shows deafult datasets names which are present in seaborn
```

```python
#how to import or use above deafult datasets 
tips=sns.load_dataset('tips')#type dataset name in paranthesis
```

```python
tips.head()
```

# Scatter chart

```python
sns.scatterplot(data=tips)
plt.show()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip') #basic chart
plt.show()
```

```python
tips['time'].unique # it shows unique values of 'time' column
```

```python
#you can show categoriacl data like this
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='time') 
plt.show()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='time',style='time') 
plt.show()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time') 
plt.show()
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='day',style='day')
plt.show()
```

# Line Ploat

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
may_flights=flights.query('month == "May"') 
sns.lineplot(data=may_flights,x='year',y='passengers')
plt.show()
```

```python
sns.lineplot(data=flights,x='year',y='passengers')
plt.show()
```

```python
flights_wide=flights.pivot('year','month','passengers')
flights_wide.head()
```

```python
flights_wide.loc[1950] 
```

```python
plt.figure(figsize=(8,8))
sns.lineplot(data=flights_wide)
plt.show()
```

```python
sns.lineplot(data=flights_wide['Aug'])
plt.show()
```

# Histogram

```python
penguins=sns.load_dataset('penguins')
penguins.head()
```

```python
sns.histplot(data=penguins,x='flipper_length_mm')
plt.show()
```

```python
sns.histplot(data=penguins,y='flipper_length_mm')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.histplot(data=penguins,x='flipper_length_mm',binwidth=1)
plt.show()
```

# Strip plot

```python
tips
```

```python
sns.stripplot(data=tips)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.stripplot(data=tips,x='day',y='total_bill')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.stripplot(data=tips,y='day',x='total_bill',jitter=0.5) # jitter is use for displacement 
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.stripplot(data=tips,x='day',y='total_bill',jitter=0.5) # jitter is use for displacement 
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.stripplot(data=tips,x='day',y='total_bill',linewidth=1.2) # jitter is use for displacement 
plt.show()
```

```python
plt.figure(figsize=(11,15))
sns.stripplot(data=tips,x='day',y='total_bill',linewidth=1.2,hue='sex') # jitter is use for displacement 
plt.show()
```
